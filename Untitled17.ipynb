```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

print("✅ Libraries loaded. Ready for data.")
```

```python
import pandas as pd

# Load the files (Assuming these are the names in your sidebar)
# If your filenames are different, change them here:
sentiment_df = pd.read_csv('fear_greed_index.csv')
trader_df = pd.read_csv('historical_data.csv')

# --- PART A: DATA AUDIT ---
def audit_data(df, name):
    print(f"--- {name} Audit ---")
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    print(f"Column Names: {list(df.columns)}")
    print(f"Missing Values:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
    print(f"Duplicates: {df.duplicated().sum()}\n")

audit_data(sentiment_df, "Sentiment Data")
audit_data(trader_df, "Trader Data")

# Show the first 2 rows to verify the data structure
print("Sentiment Preview:")
print(sentiment_df.head(2))
print("\nTrader Data Preview:")
print(trader_df.head(2))
```

```python
# 1. Convert Sentiment 'date' to standard date format
sentiment_df['Date'] = pd.to_datetime(sentiment_df['date']).dt.date

# 2. Convert Trader 'Timestamp IST' to standard date format
# We split the string to ignore the time (HH:MM) and just keep the Day
trader_df['Date'] = pd.to_datetime(trader_df['Timestamp IST'], dayfirst=True).dt.date

# 3. Merge the datasets on the new 'Date' column
merged_df = pd.merge(trader_df, sentiment_df, on='Date', how='inner')

print(f"✅ Success! Merged Rows: {len(merged_df)}")
print(f"Combined Columns: {list(merged_df.columns)}")
```

```python
# Group by Account and Date to get daily performance
daily_performance = merged_df.groupby(['Account', 'Date', 'classification']).agg(
    daily_pnl=('Closed PnL', 'sum'),
    trade_count=('Account', 'count'),
    avg_size_usd=('Size USD', 'mean')
).reset_index()

# Calculate 'is_win' (1 if daily profit > 0, else 0)
daily_performance['is_win'] = (daily_performance['daily_pnl'] > 0).astype(int)

print("✅ Daily performance metrics calculated.")
print(daily_performance.head())
```

```python
# 1. Performance Comparison
performance_stats = daily_performance.groupby('classification').agg(
    avg_pnl=('daily_pnl', 'mean'),
    win_rate=('is_win', 'mean'),
    avg_trades_per_day=('trade_count', 'mean')
).reset_index()

# Convert win_rate to percentage for easier reading
performance_stats['win_rate'] = (performance_stats['win_rate'] * 100).round(2)

print("--- Performance Analysis by Sentiment ---")
print(performance_stats)

# 2. Visualizing the results
plt.figure(figsize=(10, 5))

# Plot Average PnL
plt.subplot(1, 2, 1)
sns.barplot(data=performance_stats, x='classification', y='avg_pnl', palette='viridis')
plt.title('Average Daily PnL by Sentiment')
plt.ylabel('Average PnL (USD)')

# Plot Win Rate
plt.subplot(1, 2, 2)
sns.barplot(data=performance_stats, x='classification', y='win_rate', palette='magma')
plt.title('Win Rate % by Sentiment')
plt.ylabel('Win Rate (%)')

plt.tight_layout()
plt.show()
```

```python
# Behavior Comparison
behavior_stats = daily_performance.groupby('classification').agg(
    avg_position_size=('avg_size_usd', 'mean'),
    avg_trades_per_day=('trade_count', 'mean')
).reset_index()

print("\n--- Behavioral Analysis by Sentiment ---")
print(behavior_stats)

# Plotting Behavior
plt.figure(figsize=(10, 5))
sns.scatterplot(data=daily_performance, x='trade_count', y='daily_pnl', hue='classification', alpha=0.5)
plt.title('Trade Frequency vs. Profitability')
plt.xlabel('Number of Trades per Day')
plt.ylabel('Daily PnL')
plt.show()
```

```python
# 1. Calculate the average trade frequency per account
account_activity = daily_performance.groupby('Account')['trade_count'].mean().reset_index()

# 2. Segment accounts into 'High Activity' and 'Low Activity'
median_trades = account_activity['trade_count'].median()
account_activity['segment'] = np.where(account_activity['trade_count'] > median_trades, 'High Activity', 'Low Activity')

# 3. Merge this segment label back into our main daily performance table
daily_performance = pd.merge(daily_performance, account_activity[['Account', 'segment']], on='Account')

# 4. Compare Win Rate by Segment and Sentiment
segment_analysis = daily_performance.groupby(['segment', 'classification'])['is_win'].mean().unstack() * 100

print(f"--- Win Rate (%) by Segment and Sentiment ---")
print(segment_analysis)

# 5. Visualize the segment comparison
segment_analysis.plot(kind='bar', figsize=(10, 6), colormap='coolwarm')
plt.title('Win Rate by Trader Segment and Market Sentiment')
plt.ylabel('Win Rate (%)')
plt.xlabel('Trader Segment')
plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

```python
# Find the most profitable sentiment for each segment
best_sentiment = daily_performance.groupby(['segment', 'classification'])['daily_pnl'].mean().idxmax()
worst_sentiment = daily_performance.groupby(['segment', 'classification'])['daily_pnl'].mean().idxmin()

print(f"INSIGHT 1: The most profitable scenario is {best_sentiment[0]} during {best_sentiment[1]}.")
print(f" INSIGHT 2: The highest risk of loss is for {worst_sentiment[0]} during {worst_sentiment[1]}.")
```
